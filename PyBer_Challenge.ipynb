### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "../PyBer_Analysis/Resources/city_data.csv"
ride_data_to_load = "../PyBer_Analysis/Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_by_city_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides_by_city_type
```

```python
# 2. Get the total drivers for each city type
drivers_count_each_city_type = city_data_df.groupby(["type"]).sum()["driver_count"]
drivers_count_each_city_type
```

```python
#  3. Get the total amount of fares for each city type
total_fare_each_city = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fare_each_city
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_per_ride = total_fare_each_city / total_rides_by_city_type
avg_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_per_driver = total_fare_each_city / drivers_count_each_city_type
avg_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
new_pyber_summary_df = pd.DataFrame()

new_pyber_summary_df["Total Rides"] = total_rides_by_city_type
new_pyber_summary_df["Total Drivers"] = drivers_count_each_city_type
new_pyber_summary_df["Total Fares"] = total_fare_each_city
new_pyber_summary_df["Average Fare Per Ride"] = avg_fare_per_ride
new_pyber_summary_df["Average Fare Per Driver"] = avg_fare_per_driver

new_pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
new_pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
new_pyber_summary_df["Total Rides"] = new_pyber_summary_df["Total Rides"].map("{:,}".format)
new_pyber_summary_df["Total Drivers"] = new_pyber_summary_df["Total Drivers"].map("{:,}".format)
new_pyber_summary_df["Total Fares"] = new_pyber_summary_df["Total Fares"].map("${:,.2f}".format)
new_pyber_summary_df["Average Fare Per Ride"] = new_pyber_summary_df["Average Fare Per Ride"].map("${:.2f}".format)
new_pyber_summary_df["Average Fare Per Driver"] = new_pyber_summary_df["Average Fare Per Driver"].map("${:.2f}".format)

new_pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
fares_by_city_date_df = pyber_data_df.groupby(['type','date']).sum().fare.to_frame()
fares_by_city_date_df.head()
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
fares_by_city_date_df = fares_by_city_date_df.reset_index()
fares_by_city_date_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
fares_by_city_date_df = fares_by_city_date_df.pivot(index='date', columns='type', values='fare')
fares_by_city_date_df
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
new_fares_date_df = fares_by_city_date_df.loc['2019-01-01':'2019-04-29']
new_fares_date_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
new_fares_date_df.index = pd.to_datetime(new_fares_date_df.index)
new_fares_date_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
new_fares_date_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
new_fares_date_df = new_fares_date_df.resample('W').sum()
new_fares_date_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

new_fares_date_df.plot(figsize=(15, 8))

# Add graph properties
plt.title("Total Fare by City Type", fontsize=16)
plt.ylabel("Fare ($USD)", fontsize=14)
plt.xlabel("Month", fontsize=14)

# Add a legend
lgnd = plt.legend(fontsize="10", loc="best", title="City Type")
lgnd.get_title().set_fontsize(12)

# Save Figure
plt.savefig("../PyBer_Analysis/Charts_Outcomes/Pyber_fare_summary.png")

plt.show()
plt.tight_layout()
```
